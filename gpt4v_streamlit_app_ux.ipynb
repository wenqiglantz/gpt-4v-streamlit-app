# Unleashing the power of GPT-4 Vision in Steamlit App Development

We will explore GPT-4 Vision and work with it as a pair programmer to develop a Streamlit app, specifically:
* Turn a design sketch into a Streamlit app
* Suggest UX enhancements
* Document the app with README markdown

```python
import os, openai, logging, sys

OPENAI_API_TOKEN = "sk-#####################"
os.environ["OPENAI_API_KEY"] = OPENAI_API_TOKEN

logging.basicConfig(stream=sys.stdout, level=logging.DEBUG)
```

```python
from llama_index.multi_modal_llms.openai import OpenAIMultiModal
from llama_index import SimpleDirectoryReader

# put your local images directory here
image_documents = SimpleDirectoryReader("./images").load_data()

llm_gpt4v = OpenAIMultiModal(
    model="gpt-4-vision-preview", api_key=OPENAI_API_TOKEN, max_new_tokens=1000
)
```

## Turn a design sketch into a Streamlit app

```python
from PIL import Image
import matplotlib.pyplot as plt

imageUrl = "./images/mockup.png"
image = Image.open(imageUrl).convert("RGB")

plt.figure(figsize=(16, 5))
plt.imshow(image)
```

```python
response = llm_gpt4v.complete(
    prompt="Create a Streamlit app from this mockup. You will use the Echarts library. The three rectangles under company logo are buttons. The table in the content area can be dataframe. Put a text paragraph under the page title.",
    image_documents=image_documents,
)
print(str(response))
```

```python
# React app code generation is not as ideal
response = llm_gpt4v.complete(
    prompt="Create a React app from this mockup. The three rectangles under company logo are buttons. The table in the content area can be dataframe. Please produce the code as close to the mockup as possible.",
    image_documents=image_documents,
)
print(str(response))
```

## Suggest UX enhancements

```python
from PIL import Image
import matplotlib.pyplot as plt

imageUrl = "./images/streamlit-app.png"
image = Image.open(imageUrl).convert("RGB")

plt.figure(figsize=(16, 5))
plt.imshow(image)
```

```python
response = llm_gpt4v.complete(
    prompt="Provide suggestions to enhance the usability of this particular Streamlit app as outlined in the image.",
    image_documents=image_documents,
)
print(str(response))
```

## Document the app

```python
response = llm_gpt4v.complete(
    prompt="Analyze the image I've uploaded. Create a README about it in Markdown syntax. Add a prerequisite on how to install Streamlit.",
    image_documents=image_documents,
)
print(str(response))
```
